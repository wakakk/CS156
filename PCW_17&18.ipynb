```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
from keras.models import Sequential
from keras.layers import Dense
```

```python
(moon_x,moon_y) = datasets.make_moons(n_samples=10000, noise=0.05, random_state=42)
(circle_x,circle_y) = datasets.make_circles(n_samples=10000, noise=0.025, random_state=42)
```

```python
XM_train, XM_test, ym_train, ym_test = train_test_split(moon_x, moon_y, test_size=0.33, random_state=42)
XC_train, XC_test, yc_train, yc_test = train_test_split(circle_x, circle_y, test_size=0.33, random_state=42)
```

```python
model = Sequential()
model.add(Dense(12, input_dim=2, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='relu'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(XM_train, ym_train, epochs=10, batch_size=10, verbose=0)
test_results = model.evaluate(XM_test, ym_test, verbose=0)
print(f'Test results - Loss: {test_results[0]} - Accuracy: {test_results[1]*100}%')
plot_decision_regions(XM_test, ym_test, clf=model, legend=2)
plt.show()
```

##Session 10.1

```python
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
from glob import glob
from random import shuffle, seed
import numpy as np
import pandas as pd

model = VGG16(weights='imagenet', include_top=False)
```

```python
# Import data and flatten each image into an array

# 206 images for each
shirt_files = glob('./shirts/*')
jersey_files = glob('./jerseys/*')
```

```python
raw_data = []


def get_raw_data(filenames, label):
    for img_path in filenames:
        img = image.load_img(img_path, target_size=(224, 224))
        img_data = image.img_to_array(img)
        img_data = np.expand_dims(img_data, axis=0)
        img_data = preprocess_input(img_data)

        vgg16_feature = model.predict(img_data)
        vgg16_feature_np = np.array(vgg16_feature)
        raw_data.append([vgg16_feature_np.flatten(), label])
        

get_raw_data(shirt_files, 0)
get_raw_data(jersey_files, 1)
```

```python
seed(10)
shuffle(raw_data)

X = np.array([cd for (cd, label) in raw_data])
y = np.array([label for (cd,label) in raw_data])

# Train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.6,random_state=42)
```

```python
original_lr = LogisticRegression(max_iter=1000, solver="liblinear")
original_lr.fit(X_train, y_train)

print("Training Score", original_lr.score(X_train, y_train))
print("Testing Score", original_lr.score(X_test, y_test))
```
